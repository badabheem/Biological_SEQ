```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('breast-cancer-data.csv')

# Preprocessing: remove 'id' and 'Unnamed: 32' (which is all NaN)
# Also encode diagnosis for inclusion in the heatmap
df_clean = df.drop(columns=['id', 'Unnamed: 32'], errors='ignore')
df_clean['diagnosis'] = df_clean['diagnosis'].map({'M': 1, 'B': 0})

# Calculate the correlation matrix for all numeric columns
corr_matrix = df_clean.corr()

# Set up the matplotlib figure
plt.figure(figsize=(20, 16))

# Draw the heatmap
# annot=True might be too cluttered for 31x31, let's use a smaller font or conditional annotation if possible.
# For now, I'll use annot=False or very small labels to keep it clean.
sns.heatmap(corr_matrix, annot=True, fmt='.1f', cmap='coolwarm', linewidths=0.5, annot_kws={"size": 8})

plt.title('Correlation Heatmap of All Features', fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

# Save the plot
plt.savefig('full_correlation_heatmap.png')

# Radius Mean vs Diagnosis
plt.figure(figsize=(8, 6))
sns.boxplot(x='diagnosis', y='radius_mean', data=df_clean)
plt.title('Radius Mean by Diagnosis')
plt.savefig('radius_mean_boxplot.png')
```

**Key Observations** from the Heatmap:

High Inter-Correlation (Multicollinearity): You can see strong red blocks ($r \approx 1.0$) between features like **radius**, **perimeter**, and **area**. This is expected because they are mathematically related (e.g., perimeter and area are calculated using the radius).

Diagnosis Correlation: The row/column for diagnosis shows which features have the strongest relationship with the tumor being malignant.

Features like **concave points_worst**, **perimeter_worst**, and **radius_worst** show deep red colors, indicating high positive correlation.

Fractal Dimension and Texture: Several "**standard error**" (_se) features and fractal_dimension show lighter or even blue colors, indicating they have a much weaker or even negative correlation with the other main size-related features.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('breast-cancer-data.csv')
df_clean = df.drop(columns=['id', 'Unnamed: 32'], errors='ignore')
df_clean['diagnosis_encoded'] = df_clean['diagnosis'].map({'M': 1, 'B': 0})

# 1. Heatmap for Multicollinearity (Size-related features)
size_features = [
    'radius_mean', 'perimeter_mean', 'area_mean',
    'radius_se', 'perimeter_se', 'area_se',
    'radius_worst', 'perimeter_worst', 'area_worst'
]
plt.figure(figsize=(10, 10))
sns.heatmap(df_clean[size_features].corr(), annot=True, cmap='Reds', fmt='.2f')
plt.title('Heatmap: Multicollinearity in Size-Related Features', fontsize=14)
plt.savefig('multicollinearity_heatmap_specific.png')
plt.show()

# 2. Heatmap for Diagnosis Correlation
# Identifying top correlated features with diagnosis to show in the heatmap
correlations = df_clean.drop(columns=['diagnosis']).corr()['diagnosis_encoded'].sort_values(ascending=False)
top_10_plus_diag = correlations.head(11).index.tolist()

plt.figure(figsize=(10, 10))
sns.heatmap(df_clean[top_10_plus_diag].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Top Features Highly Correlated with Diagnosis', fontsize=14)
plt.savefig('diagnosis_top_correlation_heatmap.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data again (or use previous df if still in memory, but safer to reload or use the clean version)
df = pd.read_csv('breast-cancer-data.csv')
df_clean = df.drop(columns=['id', 'Unnamed: 32'])

# Top 4 features from the previous correlation analysis
top_features = ['concave points_worst', 'perimeter_worst', 'concave points_mean', 'radius_worst']

# 1. Distribution plots for top features
# Set up the matplotlib figure
plt.figure(figsize=(8, 6))
for i, feature in enumerate(top_features, 1):
    plt.subplot(2, 2, i)
    sns.kdeplot(data=df_clean, x=feature, hue='diagnosis', fill=True, palette='magma')
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.savefig('top_features_distribution.png')

# 2. Scatter plot of the two most correlated features
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_clean, x='perimeter_worst', y='concave points_worst', hue='diagnosis', style='diagnosis', palette='viridis', alpha=0.7)
plt.title('Perimeter Worst vs Concave Points Worst')
plt.savefig('scatter_top_features.png')

# 3. Simple Feature Importance via Random Forest (just for insight)
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

X = df_clean.drop(columns=['diagnosis'])
y = LabelEncoder().fit_transform(df_clean['diagnosis']) # M=1, B=0

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': rf.feature_importances_
}).sort_values(by='importance', ascending=False)

print("Top 10 Feature Importances:")
print(feature_importance.head(10))
```

The distribution plots (KDE) for the top 4 features—concave points_worst, perimeter_worst, concave points_mean, and radius_worst—show a very clear separation between the two classes.

Malignant (M) cases consistently exhibit higher values across all these metrics.

There is very little overlap in the distributions of concave points_worst and concave points_mean, making them excellent candidates for classification.

The scatter plot of Perimeter Worst vs. Concave Points Worst highlights a strong linear-like trend where Malignant cases cluster in the upper right. This indicates that as the tumor gets larger (perimeter) and more irregular (concave points), the likelihood of it being malignant increases significantly.

To quantify the predictive power, Random Forest model to rank features by their "importance" in making a correct classification:

area_worst: 13.9% importance

concave points_worst: 13.2% importance

concave points_mean: 10.7% importance

radius_worst: 8.2% importance

perimeter_worst: 8.0% importance

These top 5 features alone account for over 54% of the model's decision-making power, confirming that the "worst" (largest/extreme) measurements are often more telling than the "mean" measurements for diagnosis.

**Conclusion**
The analysis confirms that breast cancer diagnosis in this dataset is highly dependent on the size (area, radius, perimeter) and shape irregularity (concavity, concave points) of the cell nuclei. Specifically, the "worst" case measurements are the most reliable indicators of malignancy.
